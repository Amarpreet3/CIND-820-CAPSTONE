```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Read the individual CSV files
btc_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_Bitcoin.csv")
eth_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_Ethereum.csv")
xrp_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_XRP.csv")
ltc_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_Litecoin.csv")
usdc_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_USDCoin.csv")

# Create a dictionary of coins and their respective dataframes
coins = {
    'Bitcoin': btc_df,
    'Ethereum': eth_df,
    'XRP': xrp_df,
    'Litecoin': ltc_df,
    'USDCoin': usdc_df
}

# Set the color palette and background color
colors = ['#1f77b4']
plt.rcParams['axes.facecolor'] = '#eef6ff'

# Create a 3x2 grid of subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# Iterate over the coins dictionary and plot the forecast in each subplot
for i, (coin, df) in enumerate(coins.items()):
    row = i // 2
    col = i % 2
    ax = axes[row, col]

    # Convert the 'Date' column to a datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Rename the 'Date' and 'Close' columns to 'ds' and 'y', respectively
    df = df.rename(columns={'Date': 'ds', 'Close': 'y'})

    # Create and fit the model
    model = Prophet(daily_seasonality=False)  # Disable daily seasonality
    model.fit(df)

    # Generate future dates for prediction
    future_dates = model.make_future_dataframe(periods=30)  # Adjust the number of future data points as needed

    # Make predictions
    forecast = model.predict(future_dates)

    # Evaluate the model
    df_eval = forecast.set_index('ds')[['yhat']].join(df.set_index('ds')['y']).reset_index()
    df_eval.fillna(df_eval.mean(), inplace=True)  # Fill missing values with mean
    mse = ((df_eval['yhat'] - df_eval['y']) ** 2).mean()
    rmse = mse ** 0.5
    mae = (df_eval['yhat'] - df_eval['y']).abs().mean()
    r2 = r2_score(df_eval['y'], df_eval['yhat'])

    # Plot the forecasted values
    ax.plot(df['ds'], df['y'], color=colors[0], label='Actual')
    ax.plot(forecast['ds'], forecast['yhat'], color='red', linestyle='dashed', label='Forecast')
    ax.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color=colors[0], alpha=0.3)

    # Customize the subplot
    ax.set_title(coin)
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.legend()

    # Display evaluation metrics
    textstr = f"MSE: {mse:.2f}\nRMSE: {rmse:.2f}\nMAE: {mae:.2f}\nR2 Score: {r2:.2f}"
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10, verticalalignment='top')

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Read the individual CSV files
btc_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_Bitcoin.csv")
eth_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_Ethereum.csv")
xrp_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_XRP.csv")
ltc_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_Litecoin.csv")
usdc_df = pd.read_csv("https://raw.githubusercontent.com/Amarpreet3/CIND-820-CAPSTONE/main/Dataset/coin_USDCoin.csv")

# Create a dictionary of coins and their respective dataframes
coins = {
    'Bitcoin': btc_df,
    'Ethereum': eth_df,
    'XRP': xrp_df,
    'Litecoin': ltc_df,
    'USDCoin': usdc_df
}

# Initialize an empty DataFrame to store the evaluation metrics
metrics_df = pd.DataFrame(columns=['MSE', 'RMSE', 'MAE', 'R2 Score'])

# Iterate over the coins dictionary and evaluate the model for each coin
for coin, df in coins.items():
    # Convert the 'Date' column to a datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Rename the 'Date' and 'Close' columns to 'ds' and 'y', respectively
    df = df.rename(columns={'Date': 'ds', 'Close': 'y'})

    # Create and fit the model
    model = Prophet()
    model.fit(df)

    # Generate future dates for prediction
    future_dates = model.make_future_dataframe(periods=30)  # Adjust the number of future data points as needed

    # Make predictions
    forecast = model.predict(future_dates)

    # Evaluate the model
    df_eval = forecast[['ds', 'yhat']].merge(df[['ds', 'y']], on='ds', how='inner').dropna()
    mse = mean_squared_error(df_eval['y'], df_eval['yhat'])
    rmse = mse ** 0.5
    mae = mean_absolute_error(df_eval['y'], df_eval['yhat'])
    r2 = r2_score(df_eval['y'], df_eval['yhat'])

    # Add the evaluation metrics to the DataFrame
    metrics_df.loc[coin] = [mse, rmse, mae, r2]

# Display the table of evaluation metrics
print(metrics_df)
```
